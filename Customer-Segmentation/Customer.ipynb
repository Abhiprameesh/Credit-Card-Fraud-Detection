```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
#Data Collection 
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
df.shape
df.info()
```

```python
df.isnull().sum()
```

```python
#choosing annual income and spending score for clustering
X = df.iloc[:, [3, 4]].values
print(X)
```

```python
#choosing the number of clusters using the elbow method
#WCSS --> Within-Cluster Sum of Squares
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
#optimum number of clusters is 5
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=0)
```

```python
Y =kmeans.fit_predict(X)
```

```python
print(Y)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X[Y == 0, 0], X[Y == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[Y == 2, 0], X[Y == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[Y == 3, 0], X[Y == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[Y == 4, 0], X[Y == 4, 1], s=100, c='magenta', label='Cluster 5')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income (k$)')    
plt.ylabel('Spending Score (1-100)')
plt.show()
```
